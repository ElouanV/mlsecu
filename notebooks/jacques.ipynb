```python
# LS 201	, LS 202	,LSL 203	,LSLL 203 ,LS 401, LSH 601,	LSH 602,	LSH 603,	LSL 601,	LSL 602,	LSL 603
columns_with_dct_string = ['LS 201', 'LS 202', 'LSL 203', 'LSLL 203', 'LS 401', 'LSH 601', 'LSH 602', 'LSH 603', 'LSL 601', 'LSL 602', 'LSL 603']
```

```python
# Apply ast.literal_eval to convert string to dictionary
import ast
for col in columns_with_dct_string:
    df[col] = df[col].apply(ast.literal_eval)
# Define a function to expand the dictionary values into columns
def expand_dictionary_values(df, column_name):
    # Get the unique keys
    unique_keys = set()
    for row in df[column_name]:
        unique_keys.update(row.keys())
    # Create a column for each key
    for key in unique_keys:
        df[key] = 0
    # Fill in the values
    # Create a column for each key
    for key in unique_keys:
        new_column_name = column_name + '_' + key
        df[new_column_name] = df[column_name].apply(lambda x: x[key] if key in x else 0)
        # Drop the original column
    df = df.drop(column_name, axis=1)
    return df

for col in columns_with_dct_string:
    df = expand_dictionary_values(df, col)
    
```

```python
# Drop columns that contains only 'Active' or 'Inactive'
for col in df.columns:
    if len(df[col].unique()) == 1 and df[col].dtype == 'object' and df[col].unique()[0] in ['Active', 'Inactive']:
        df = df.drop(col, axis=1)
```

```python
# Turn columns which values are only Active or Inactive into bool
for col in df.columns:
    if len(df[col].unique()) == 2 and df[col].dtype == 'object':
        df[col] = df[col].apply(lambda x: True if x == 'Active' else False)
```

```python
for col in df.columns:
    if len(df[col].unique()) == 1 and df[col].dtype in ['bool', 'int64']:
        df = df.drop(col, axis=1)
```

```python
# COnvert columns that can be converted to int
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = df[col].astype('float64')
        except:
            print(f'Cannot convert {col} to int')
            pass
```
